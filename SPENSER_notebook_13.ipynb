```python
#List containing the range of ages
age_range_list = [(16,24),(25,49),(50,64)]

# Probability of NSSEC by highest degree and age range
for age_range in age_range_list:
    tot_no_qualif = 0
    tot_below = 0
    tot_higher = 0
    for occ in range(1,10):
        tot_no_qualif = tot_no_qualif + df_byocc[str(age_range[0])+'_'+str(age_range[1])+'_occ_'+str(occ)+'_no_qualif']
        tot_below = tot_below + df_byocc[str(age_range[0])+'_'+str(age_range[1])+'_occ_'+str(occ)+'_below']
        tot_higher = tot_higher + df_byocc[str(age_range[0])+'_'+str(age_range[1])+'_occ_'+str(occ)+'_higher']
        
    for occ in range(1,10):
        (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_no_qualif"]) = df_byocc[str(age_range[0])+'_'+str(age_range[1])+'_occ_'+str(occ)+'_no_qualif']/tot_no_qualif
        (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_below"]) = df_byocc[str(age_range[0])+'_'+str(age_range[1])+'_occ_'+str(occ)+'_below']/tot_below
        (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_higher"]) = df_byocc[str(age_range[0])+'_'+str(age_range[1])+'_occ_'+str(occ)+'_higher']/tot_higher
    
# Add probability of NSSEC nan by highest degree and age range
for age_range in age_range_list:
    min_no_qualif = 1.0
    min_below = 1.0
    min_higher = 1.0
    for occ in range(1,10):
        if(globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_no_qualif"].values[0]<min_no_qualif):
            min_no_qualif = (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_no_qualif"].values[0])
        if((globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_below"].values[0])<min_below):
            min_below = (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_below"].values[0])
        if((globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_higher"].values[0])<min_higher):
            min_higher = (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_higher"].values[0])
    for occ in range(1,10):
        (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_no_qualif"]) = (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_no_qualif"]) - min_no_qualif/10.0
        (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_below"]) = (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_below"])-min_below/10.0
        (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_higher"]) = (globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_higher"])-min_higher/10.0
        
    (globals()[f"proba_{age_range[0]}_{age_range[1]}_nan_no_qualif"]) = min_no_qualif
    (globals()[f"proba_{age_range[0]}_{age_range[1]}_nan_below"]) = min_below
    (globals()[f"proba_{age_range[0]}_{age_range[1]}_nan_higher"]) = min_higher
```

```python
## LIST OF OA_AREAS that has been generated before
# Create a list with all OA unique ID values
AreaOA_list = df_persons_NE_16_64['Area_OA_x'].tolist()
# Remove duplicates
AreaOA_list =  list(set(AreaOA_list))


#Create an empty list where the small blocks of dataframes of NO QUALIF will be stored
persons_NE_no_qualif_chosen_all_list = []

#List containing the range of ages
age_range_list = [(16,24),(25,49),(50,64)]

# INITIAL values to transform data from 2011 to 2021 based on the relationship (2021/2011)
# These values will be updated everytime an iteration is not within +-2% of the value of 2021
#VALUES IT 0
#no_qualif_conversor_16_24 = 0.311
#no_qualif_conversor_25_49 = 0.723
#no_qualif_conversor_50_64 = 0.375
#VALUES IT 1
no_qualif_conversor_16_24 = 0.311
no_qualif_conversor_25_49 = 0.6
no_qualif_conversor_50_64 = 0.3

for age_range in age_range_list:

    # Set the percentage of no qualif range of age to zero at the begining.
    (globals()[f"total_{age_range[0]}_{age_range[1]}_no_qualif_percentage"]) = 0

    # let's start assigning highest qualification to the persons in the synthetic population
    ## Based on the OA level, age
    print('CONVERSOR Value IS', (globals()[f"no_qualif_conversor_{age_range[0]}_{age_range[1]}"]))

    # Clear the list everytime there is a need for a new iteration
    persons_NE_no_qualif_chosen_all_list.clear()

    # Clear the dataframe everytime there is a need for an new iteration          
    globals()[f"df_{age_range[0]}_{age_range[1]}_no_qualif"] = pd.DataFrame()         


    #Create a variable that counts the number of OA areas iterated
    OA_area_counter = 0

    for OA_area in AreaOA_list:
        OA_area_counter += 1
        print("Number of OA areas in iteration: ", (OA_area_counter, len(AreaOA_list)))

        #Select the row of the df_LC5102EW that is related to the selected OA area:
        df_LC5102EW_OAarea = df_LC5102EW.loc[(df_LC5102EW['geography'] == OA_area)] 

        # Select those people that live in the selected OA area (people already between 16 and 64 years old):
        df_persons_NE_16_64_OAarea = df_persons_NE_16_64.loc[(df_persons_NE_16_64['Area_OA_x'] == OA_area)]

        # The NO QUALIF people will be chosen from this dataset.
        #The number of them will depend on the value of variable "globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_2021"]"
        globals()[f"df_{age_range[0]}_{age_range[1]}"] = df_persons_NE_16_64_OAarea.loc[(df_persons_NE_16_64_OAarea['Age'] >= age_range[0]) & (df_persons_NE_16_64_OAarea['Age'] <= age_range[1])]

        # Calculate the total number of people with the specific range of age living in that OA area:
        globals()[f"total_{age_range[0]}_{age_range[1]}_in_OA_2021"] = len(globals()[f"df_{age_range[0]}_{age_range[1]}"])

        #Create a variable per type of qualification level based on range of age
        col_age0_age1_no_qualif = str(age_range[0]) + "_" + str(age_range[1]) + "_no_qualif"
        col_age0_age1_below = str(age_range[0]) + "_" + str(age_range[1]) + "_below"
        col_age0_age1_higher = str(age_range[0]) + "_" + str(age_range[1]) + "_higher"


        # Identify the number of people based on range of age in the selected OA area that have no qualif
        # Value from 2011!!
        # Value comming table LC5102EW that has to be updated to 2021
        globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_LC5102EW"] = df_LC5102EW_OAarea.iloc[0,df_LC5102EW_OAarea.columns.get_loc(col_age0_age1_no_qualif)]

        # Identify the number of people based on range of age in the selected OA area that have below qualif
        # Value from 2011!!
        # Value comming table LC5102EW that has to be updated to 2021
        globals()[f"below_{age_range[0]}_{age_range[1]}_LC5102EW"] = df_LC5102EW_OAarea.iloc[0,df_LC5102EW_OAarea.columns.get_loc(col_age0_age1_below)]

        # Identify the number of people based on range of age in the selected OA area that have higher qualif
        # Value from 2011!!
        # Value comming table LC5102EW that has to be updated to 2021
        globals()[f"higher_{age_range[0]}_{age_range[1]}_LC5102EW"] = df_LC5102EW_OAarea.iloc[0,df_LC5102EW_OAarea.columns.get_loc(col_age0_age1_higher)]
        
        # Value with the total number of people living in the seleted OA area (table LC5102EW)
        globals()[f"Total_{age_range[0]}_{age_range[1]}_LC5102EW"] = globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_LC5102EW"] + globals()[f"below_{age_range[0]}_{age_range[1]}_LC5102EW"] + globals()[f"higher_{age_range[0]}_{age_range[1]}_LC5102EW"]


        
        # Ratio of people 2021 vs 2011:
        ## If the number of people of the selected range of age is greater than 0 (e.g., there are people in the OA with this specific range of age)
        if (globals()[f"Total_{age_range[0]}_{age_range[1]}_LC5102EW"]) > 0:
            globals()[f"ratio_people_{age_range[0]}_{age_range[1]}_2021_2011"] = (globals()[f"total_{age_range[0]}_{age_range[1]}_in_OA_2021"]/globals()[f"Total_{age_range[0]}_{age_range[1]}_LC5102EW"])
            ## If value > 1, it means there are more people in 2021 in the selected OA area of the specific ange of age
                    
        # If there is no people (age range) in the OA area, then the ratio will be equal to 1
        else:
            globals()[f"ratio_people_{age_range[0]}_{age_range[1]}_2021_2011"] = 1
                    
                
        # New value for 2021 =  (Value from 2011 (table LC5102EW)) * (no_qualif conversor value (based on age range)) * (ratio of people 2021 vs 2011 (c)) 
        ## This value is the number of people that will be randomly assigned "no qualif" based on their OA area, range of age
        globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_2021"] =  int(round(globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_LC5102EW"] * globals()[f"no_qualif_conversor_{age_range[0]}_{age_range[1]}"] * globals()[f"ratio_people_{age_range[0]}_{age_range[1]}_2021_2011"],0)) 
                     
        print('value from census 2011')
        print(globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_LC5102EW"])
        print('value for 2021')
        print(globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_2021"])
        
        #If no_qualif people to select in this OA and age range
        if (globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_2021"] != 0):
            # If the total number of people with the specific range of age living in that OA area > people to be selected
            if (globals()[f"total_{age_range[0]}_{age_range[1]}_in_OA_2021"]) > globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_2021"]:
                # Select randomly
                selection = pd.DataFrame()
                while(len(selection.index) < globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_2021"]):
                    for occ in range(1,10):
                        people = (globals()[f"df_{age_range[0]}_{age_range[1]}"]).loc[globals()[f"df_{age_range[0]}_{age_range[1]}"]['NSSEC'] == float(occ)]
                        # Remove already selected
                        people_plus = (pd.concat([people, selection]))
                        people = people_plus.drop_duplicates(subset='PID_AreaMSOA', keep = False)
                        people = people.loc[people['NSSEC'] == float(occ)]
                        
                        if(len(people.index) != 0):
                            nb = min(len(people.index),m.ceil(globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_2021"]*globals()[f"proba_{age_range[0]}_{age_range[1]}_{occ}_no_qualif"].values[0]))
                            selection = pd.concat([selection,people.sample(nb)])

                    people = (globals()[f"df_{age_range[0]}_{age_range[1]}"]).loc[globals()[f"df_{age_range[0]}_{age_range[1]}"]['NSSEC'].isna()]
                   
                    # Remove already selected
                    people_plus = (pd.concat([people, selection]))
                    people = people_plus.drop_duplicates(subset='PID_AreaMSOA', keep = False)
                    people = people.loc[people['NSSEC'].isna()]
                    
                    if(len(people.index) != 0):
                        nb = min(len(people.index),m.ceil(globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_2021"]*globals()[f"proba_{age_range[0]}_{age_range[1]}_nan_no_qualif"]))
                        selection = pd.concat([selection,people.sample(nb)])
                no_qualif = selection.sample(globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_2021"])
            else:
                # Select all people 
                no_qualif = (globals()[f"df_{age_range[0]}_{age_range[1]}"])

            persons_NE_no_qualif_chosen_all_list.append(no_qualif)  
                
    if persons_NE_no_qualif_chosen_all_list:
        globals()[f"df_{age_range[0]}_{age_range[1]}_no_qualif"] = pd.concat(persons_NE_no_qualif_chosen_all_list, axis=0, ignore_index=True)
        
    # Calculate the TOTAL number of people with the range of age:
    globals()[f"total_{age_range[0]}_{age_range[1]}"] = len(df_persons_NE_16_64.loc[(df_persons_NE_16_64['Age'] >= age_range[0]) & (df_persons_NE_16_64['Age'] <= age_range[1])])

    #Calculate the % of people NO QUALIF with the same range of age: 
    globals()[f"total_{age_range[0]}_{age_range[1]}_no_qualif_percentage"] = (len(globals()[f"df_{age_range[0]}_{age_range[1]}_no_qualif"])/globals()[f"total_{age_range[0]}_{age_range[1]}"]) * 100
    print('Value obtained was', (globals()[f"total_{age_range[0]}_{age_range[1]}_no_qualif_percentage"]))
                       
    
```

```python
## LIST OF OA_AREAS that has been generated before
# Create a list with all OA unique ID values
AreaOA_list = df_persons_NE_65_120_qualif['Area_OA_x'].tolist()
# Remove duplicates
AreaOA_list =  list(set(AreaOA_list))

#Create an empty list where the small blocks of dataframes of HIGHER will be stored
persons_NE_higher_chosen_all_list = []

age_range = (65,120)

#List containing the range of ages
age_range_list = [(65,68),(69,73),(74,78),(79,83),(84,88),(89,93),(94,98),(99,120)]

# Apply 50-64 conversor because we don't have better
higher_conversor_65_120 = higher_conversor_50_64


# Set the percentage of higher per range of age to zero at the begining.
(globals()[f"total_{age_range[0]}_{age_range[1]}_higher_percentage"]) = 0

# let's start assigning highest qualification to the persons in the synthetic population
## Based on the OA level, age
print('CONVERSOR Value IS', (globals()[f"higher_conversor_{age_range[0]}_{age_range[1]}"]))

# Clear the list everytime there is a need for a new iteration
persons_NE_higher_chosen_all_list.clear()

# Clear the dataframe everytime there is a need for an new iteration          
globals()[f"df_{age_range[0]}_{age_range[1]}_higher"] = pd.DataFrame()         


#Create a variable that counts the number of OA areas iterated
OA_area_counter = 0

for OA_area in AreaOA_list:
    OA_area_counter += 1
    print("Number of OA areas in iteration: ", (OA_area_counter, len(AreaOA_list)))

    #Select the row of the df_LC5102EW that is related to the selected OA area:
    df_LC5102EW_OAarea = df_LC5102EW.loc[(df_LC5102EW['geography'] == OA_area)] 

    # Select those people that live in the selected OA area (people already between 65 and 120 years old):
    df_persons_NE_65_120_OAarea = df_persons_NE_65_120_qualif.loc[(df_persons_NE_65_120_qualif['Area_OA_x'] == OA_area)]

    # The higher people will be chosen from this dataset.
    #The number of them will depend on the value of variable "globals()[f"higher_{age_range[0]}_{age_range[1]}_2021"]"
    globals()[f"df_{age_range[0]}_{age_range[1]}"] = df_persons_NE_65_120_OAarea.loc[(df_persons_NE_65_120_OAarea['Age'] >= age_range[0]) & (df_persons_NE_65_120_OAarea['Age'] <= age_range[1])]

    # Calculate the total number of people with the specific range of age living in that OA area:
    globals()[f"total_{age_range[0]}_{age_range[1]}_in_OA_2021"] = len(globals()[f"df_{age_range[0]}_{age_range[1]}"])

    #Create a variable per type of qualification level based on range of age
    col_age0_age1_no_qualif = str(age_range[0]) + "_" + str(age_range[1]) + "_no_qualif"
    col_age0_age1_below = str(age_range[0]) + "_" + str(age_range[1]) + "_below"
    col_age0_age1_higher = str(age_range[0]) + "_" + str(age_range[1]) + "_higher"


    # Identify the number of people based on range of age in the selected OA area that have no qualif
    # Value from 2011!!
    # Value comming table LC5102EW that has to be updated to 2021
    globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_LC5102EW"] = df_LC5102EW_OAarea.iloc[0,df_LC5102EW_OAarea.columns.get_loc(col_age0_age1_no_qualif)]

    # Identify the number of people based on range of age in the selected OA area that have below qualif
    # Value from 2011!!
    # Value comming table LC5102EW that has to be updated to 2021
    globals()[f"below_{age_range[0]}_{age_range[1]}_LC5102EW"] = df_LC5102EW_OAarea.iloc[0,df_LC5102EW_OAarea.columns.get_loc(col_age0_age1_below)]

    # Identify the number of people based on range of age in the selected OA area that have higher qualif
    # Value from 2011!!
    # Value comming table LC5102EW that has to be updated to 2021
    globals()[f"higher_{age_range[0]}_{age_range[1]}_LC5102EW"] = df_LC5102EW_OAarea.iloc[0,df_LC5102EW_OAarea.columns.get_loc(col_age0_age1_higher)]

    # Value with the total number of people living in the seleted OA area (table LC5102EW)
    globals()[f"Total_{age_range[0]}_{age_range[1]}_LC5102EW"] = globals()[f"no_qualif_{age_range[0]}_{age_range[1]}_LC5102EW"] + globals()[f"below_{age_range[0]}_{age_range[1]}_LC5102EW"] + globals()[f"higher_{age_range[0]}_{age_range[1]}_LC5102EW"]


    # Ratio of people 2021 vs 2011:
    ## If the number of people of the selected range of age is greater than 0 (e.g., there are people in the OA with this specific range of age)
    if (globals()[f"Total_{age_range[0]}_{age_range[1]}_LC5102EW"]) > 0:
        globals()[f"ratio_people_{age_range[0]}_{age_range[1]}_2021_2011"] = (globals()[f"total_{age_range[0]}_{age_range[1]}_in_OA_2021"]/globals()[f"Total_{age_range[0]}_{age_range[1]}_LC5102EW"])
        ## If value > 1, it means there are more people in 2021 in the selected OA area of the specific ange of age

    # If there is no people (age range) in the OA area, then the ratio will be equal to 1
    else:
        globals()[f"ratio_people_{age_range[0]}_{age_range[1]}_2021_2011"] = 1


    # New value for 2021 =  (Value from 2011 (table LC5102EW)) * (higher conversor value (based on age range)) * (ratio of people 2021 vs 2011 (c)) 
    ## This value is the number of people that will be randomly assigned "higher" based on their OA area, range of age
    globals()[f"higher_{age_range[0]}_{age_range[1]}_2021"] =  int(round(globals()[f"higher_{age_range[0]}_{age_range[1]}_LC5102EW"] * globals()[f"higher_conversor_{age_range[0]}_{age_range[1]}"] * globals()[f"ratio_people_{age_range[0]}_{age_range[1]}_2021_2011"],0)) 

    print('value from census 2011')
    print(globals()[f"higher_{age_range[0]}_{age_range[1]}_LC5102EW"])
    print('value for 2021')
    print(globals()[f"higher_{age_range[0]}_{age_range[1]}_2021"])

    #If higher people to select in this OA and age range
    if (globals()[f"higher_{age_range[0]}_{age_range[1]}_2021"] != 0):
        # If the total number of people with the specific range of age living in that OA area > people to be selected
        if (globals()[f"total_{age_range[0]}_{age_range[1]}_in_OA_2021"]) > globals()[f"higher_{age_range[0]}_{age_range[1]}_2021"]:
            # Select randomly
            selection = pd.DataFrame()
            while(len(selection.index) < globals()[f"higher_{age_range[0]}_{age_range[1]}_2021"]):
                age_index = 0
                for age in age_range_list:
                    people = (globals()[f"df_{age_range[0]}_{age_range[1]}"]).loc[(globals()[f"df_{age_range[0]}_{age_range[1]}"]['Age'] >= age[0]) & (globals()[f"df_{age_range[0]}_{age_range[1]}"]['Age'] <= age[1])]
                    
                    # Remove already selected
                    people_plus = (pd.concat([people, selection]))
                    people = people_plus.drop_duplicates(subset='PID_AreaMSOA', keep = False)
                    people = people.loc[(people['Age'] >= age[0]) & (people['Age'] <= age[1])]
                        
                    if(len(people.index) != 0):
                        for gender in [1,2]:
                            people_ = people.loc[people['Sex'] == gender]
                            if(len(people_.index) != 0):
                                if(gender==1):
                                    proba = men_higher_percentage_list[age_index]
                                else:
                                    proba = women_higher_percentage_list[age_index]
                                nb = min(len(people_.index),m.ceil(globals()[f"higher_{age_range[0]}_{age_range[1]}_2021"]*proba))
                                selection = pd.concat([selection,people_.sample(nb)])
                    age_index = age_index+1
            higher = selection.sample(globals()[f"higher_{age_range[0]}_{age_range[1]}_2021"])
        else:
            # Select all people 
            higher = (globals()[f"df_{age_range[0]}_{age_range[1]}"])

        persons_NE_higher_chosen_all_list.append(higher)  

if persons_NE_higher_chosen_all_list:
    globals()[f"df_{age_range[0]}_{age_range[1]}_higher"] = pd.concat(persons_NE_higher_chosen_all_list, axis=0, ignore_index=True)

# Calculate the TOTAL number of people with the range of age:
globals()[f"total_{age_range[0]}_{age_range[1]}"] = len(df_persons_NE_65_120.loc[(df_persons_NE_65_120['Age'] >= age_range[0]) & (df_persons_NE_65_120['Age'] <= age_range[1])])

#Calculate the % of people higher with the same range of age: 
globals()[f"total_{age_range[0]}_{age_range[1]}_higher_percentage"] = (len(globals()[f"df_{age_range[0]}_{age_range[1]}_higher"])/globals()[f"total_{age_range[0]}_{age_range[1]}"]) * 100
print('Value obtained was', (globals()[f"total_{age_range[0]}_{age_range[1]}_higher_percentage"]))
```
